## This is the reviewing of mi band 4 product on amazon

```python
#importing some important libraries
import requests
import os
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
name = []
veri = []
ver = []
semi = []
s_rev = []
users_ = []
final_rat = []
user_review = []
rat_users = []
small_rev = []
n = 1
val = 1
rt = 1
vrt = 1
count_ = 0
count = 0
   
for ur in range(1,160):
    url = f'https://www.amazon.in/Mi-Smart-Band-Waterproof-up/product-reviews/B07WLL998K/ref=cm_cr_arp_d_paging_btm_next_2?ie=UTF8&reviewerType=all_reviews&pageNumber={ur}'
    page = requests.get(url)
    sp = BeautifulSoup(page.content)
    if page.status_code == 200:
        parent_class = sp.find_all('div',class_= 'a-section review aok-relative')
        for sub in zip([iter(parent_class)]):
            names = (sp.find_all('span',class_= 'a-profile-name'))
            for i in range(0,len(names)):
                name.append(names[i].get_text())
                count += 1
            else:
                val = int(count/12)
                n = int(val*2)
                users_ = name[n:]
            reviews = sp.find_all('span',class_ = 'a-size-base review-text review-text-content')
            for i in range(0,len(reviews)):
                user_review.append(reviews[i].get_text().strip())
            ratings = sp.find_all('span',class_='a-icon-alt')
            for i in range(0,len(ratings)):
                rat_users.append(ratings[i].get_text().replace(' out of 5 stars',''))
                count_ +=1
            else:
                vrt = int(count/12)
                rt = int(vrt*3)
                final_rat = rat_users[rt:]
            revi = sp.find_all('a',class_ = 'a-size-base a-link-normal review-title a-color-base review-title-content a-text-bold')
            for i in range(0,len(revi)):
                small_rev.append(revi[i].get_text().strip())
                veri = sp.find_all('span','a-size-mini a-color-state a-text-bold')
            for i in range(0,len(veri)):
                ver.append(veri[i].get_text().strip())
            s_rev = sp.find_all("a",class_="a-size-base a-link-normal review-title a-color-base review-title-content a-text-bold")
            for i in range(0,len(s_rev)):
                semi.append(s_rev[i].get_text().strip())
else:
        print('Fetched data Successfully..!')
```

## Creating DataFrame

```python
import pandas as pd
```

```python
df = pd.DataFrame({
    
    'Name' : users_,
    'Small Reviws' : semi,
    'Verified User?' : ver,
    'Reviews' : user_review,
    'Ratings' : final_rat
})
```
